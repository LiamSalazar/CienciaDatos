```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 20)
Y = x**2 + 10 * np.random.rand(20) # Generar datos con ruido (la suma es el ruido)

n = len(Y) 
m = 4
X = np.ones((n,m))
for i in range(0, n):
    for j in range(0,m):
        X[i,j] = pow(x[i], j)

Xt = np.transpose(X)
XtX_i = np.linalg.inv(np.dot(Xt, X))
B = np.dot(np.dot(XtX_i, Xt), Y)

print(B)
y2 = []
for i in range(0, len(x)):
    y2.append(B[0] + B[1]*x[i] + B[2]*x[i]**2 + B[3]*x[i]**3)

plt.scatter(x, Y, color='orange')
plt.plot(x, y2, '--')
```
