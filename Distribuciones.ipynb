Crear una funci칩n que genere una distribuci칩n gaussiana tomando como entradas la media y la desviaci칩n est치ndar.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def media(datos: list)->float:
    media = 0
    for dato in datos:
        media += dato
    return media/ len(datos)

def desviacion_estandar(datos: list)->float:
    media_datos = media(datos)
    suma = 0
    for dato in datos:
        suma += (dato - media_datos)**2
    return (suma/len(datos))**0.5

def normal(media: float, desviacion: float, x: float)->float:
    return (1/(desviacion * (2*math.pi)**0.5)) * math.e**(-0.5 * ((x - media)/desviacion)**2)

xs = np.linspace(-5, 5, 100)
plt.plot(xs, normal(0, 1,  xs), '-', label='N(0,1)')
plt.plot(xs, normal(0, 0.5, xs), '--', label='N(0,0.5)')
plt.plot(xs, normal(0, 2,  xs), ':', label='N(0,2)')
plt.plot(xs, normal(-2, 1, xs), '-.', label='N(1,1)')
plt.legend()
```

## Distribuci칩n Uniforme:
Para esta distribuci칩n todos los eventos o elementos de una variable poseen la misma probabilidad (equiprobabilidad).


Por ejemplo, los posibles resultados de tirar un dado siguen a una distribuci칩n uniforme.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.random.rand(100000000)
plt.hist(y, bins=10)
```

# Distribuci칩n exponencial:
$$
f(x) = \lambda*e^{\lambda*x}
$$
Por ejemplo, el intervalo de tiempo entre terremotos de una determinada magnitud siguen una distribuci칩n exponencial.
El tiempo entre dos potenciales de acci칩n neuronales muestra una distribuci칩n exponencial.

```python
import matplotlib.pyplot as plt
import numpy as np

l = .01
aleatorios = np.random.exponential(l, 10000)
plt.hist(aleatorios)
plt.ylabel('Frecuencia')
#plt.yscale('log')
plt.xlabel('Valor')
plt.title('Distribuci칩n Exponencial con 풭 = 0.01')
plt.show()
```
