```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3,figsize=(6,2))

a = np.zeros((10,10))
for i in range(10):
    a[i,:]=i
axes[0].imshow(a,cmap = 'Grays')

b = np.zeros((10,10))
for j in range(10):
    b[:,j]=j
axes[1].imshow(b,cmap = 'Grays')


c = np.dot(a,b)
axes[2].imshow(b)
axes[2].imshow(c, cmap = 'Grays')

fig, axes = plt.subplots(1, 3,figsize=(6,2))

a = np.zeros((10,10))
for i in range(10):
    distanciaY = (len(a[0])//2)-i
    a[i,:]=(distanciaY)**2
for i in range(10):
    distanciaX = (len(a)//2)-i
    a[:,i]+=(distanciaX)**2
    a[:,i]=a[:,i]**(1/2)
axes[0].imshow(a,cmap='Grays')

b = np.zeros((10,10))
for i in range(10):
    distanciaY = (len(a[0])//2)-i
    b[i,:]=(distanciaY)**2
for i in range(10):
    distanciaX = (len(b)//2)-i
    b[:,i]+=(distanciaX)**2
    b[:,i]=-(b[:,i]**(.8))
axes[1].imshow(b,cmap='Grays')


c = np.dot(a,b)
axes[2].imshow(c, cmap = 'Grays')
```
