```python
import matplotlib.pyplot as plt
import numpy as np

def graficar_circulos(a: float, b: float, radios: list, colors: list): # Definición de la función
    fig, ax = plt.subplots(figsize=(4,4))   # Creo una figura para graficar círculos sobre el eje ax
    
    for r, color in zip(radios, colors): # Iteración sobre arreglos de radios y colores para que haya variaciones
        x = np.linspace(a-r, a+r, 400) # Se establece para cada círculo un recorrido en x de extremo a extremo
        # Se generan los valores para y superior e inferior (raíz positiva y negativa)
        y1 = (r**2 - (x - a)**2)**(1/2) + b 
        y2 = -(r**2 - (x - a)**2)**(1/2) + b 
        # Se dibujan ambas mitades del círculo (ambas y's)
        ax.plot(x, y1, color=color) 
        ax.plot(x, y2, color=color) 
    return fig # Devuelve la figura completa

fig = graficar_circulos(0, 0, [1,2,3,4,5], colors=['blue','red','black','yellow','gray']) # Llama a la función
plt.show() # Muestra la figura completa

```
