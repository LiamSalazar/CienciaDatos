```python
import matplotlib.pyplot as plt
import numpy as np

def graficar_circulos(a: float, b: float, radios: list, colors: list): # Definición de la función
    fig, ax = plt.subplots(figsize=(4,4))   # Creo una variable fig que contendrá las variables ax para los círculos
    
    for r, color in zip(radios, colors): # Iteración sobre arreglo de radios y colores para que haya variaciones
        x = np.linspace(a-r, a+r, 400) # Se establece para cada círculo un recorrido en x de extremo a extremo
        y1 = (r**2 - (x - a)**2)**(1/2) + b # Se obtiene la parte superior del círculo en y
        y2 = -(r**2 - (x - a)**2)**(1/2) + b # Se obtiene la parte inferior del círculo en y
        ax.plot(x, y1, color=color) # Genera la gráfica de la mitad superior
        ax.plot(x, y2, color=color) # Genera la gráfica de la mitad inferior
    return fig # Devuelve la figura completa para graficar todos los círculos

fig = graficar_circulos(0, 0, [1,2,3,4,5], colors=['blue','red','black','yellow','gray']) # Llama a la función
plt.show() # Muestra la figura completa

```
