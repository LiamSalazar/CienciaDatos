```python
import matplotlib.pyplot as plt
import numpy as np

def graficar_circulo(a: float, b: float, r: float, color):
    plt.figure(figsize=(4,4)) 
    x = np.linspace(a-r,a+r,300)
    y1 = (((r**2)-(x-a)**2)**(1/2))+b
    y2 = (-((r**2)-(x-a)**2)**(1/2))+b
    plt.plot(x,y1,color=color)
    plt.plot(x,y2,color=color)
    plt.show()
graficar_circulo(0,0,2, 'blue')
```
