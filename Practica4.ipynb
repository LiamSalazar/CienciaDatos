```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2,3,figsize=(12,8))
iris = pd.read_csv("./iris.csv")
print((iris.shape))
pw = iris["petal.width"]
pl = iris["petal.length"]
sw = iris["sepal.width"]
sl = iris["sepal.length"]

# Grfica 1
axes[0][0].scatter(sl, sw)
axes[0][0].set_ylabel("Ancho del spalo")
axes[0][0].set_xlabel("Longitud del spalo")

# Grfica 2
axes[0][1].scatter(sl, pl)
axes[0][1].set_ylabel("Longitud del ptalo")
axes[0][1].set_xlabel("Longitud del spalo")

# Grfica 3
axes[0][2].scatter(sl, pw)
axes[0][2].set_ylabel("Ancho del ptalo")
axes[0][2].set_xlabel("Longitud del spalo")

# Grfica 4
axes[1][0].scatter(sw, pl)
axes[1][0].set_ylabel("Longitud del ptalo")
axes[1][0].set_xlabel("Ancho del spalo")

# Grfica 5
axes[1][1].scatter(sw, pw)
axes[1][1].set_ylabel("Ancho del ptalo")
axes[1][1].set_xlabel("Ancho del spalo")

# Grfica 6
axes[1][2].scatter(pw, pl)
axes[1][2].set_ylabel("Longitud del ptalo")
axes[1][2].set_xlabel("Ancho del ptalo")

plt.show()
```
