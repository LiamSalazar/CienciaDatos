```python
import pandas as pd
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2,3,figsize=(12,8)) # Creo la figura con las 6 grficas que har
iris = pd.read_csv("./iris.csv") # Leo el archivo CSV con pandas
# Guardo en arreglos las columnas del DataFrame
pw = iris["petal.width"]
pl = iris["petal.length"]
sw = iris["sepal.width"]
sl = iris["sepal.length"]

# Asignacin de colores por especie
color_map = {'Setosa': 'skyblue', 'Versicolor': 'purple', 'Virginica': 'violet'} # Mapeo de colores por especie mediante un diccionario
iris['color'] = iris["variety"].map(color_map) # Creo una nueva columna en el DataFrame con los colores de cada especie

# Grfica 1
axes[0][0].scatter(sl, sw, c=iris['color']) # Asignacin de eje x, eje y y color
axes[0][0].set_ylabel("Ancho del spalo")
axes[0][0].set_xlabel("Longitud del spalo")

# Grfica 2
axes[0][1].scatter(sl, pl, c=iris['color']) # Asignacin de eje x, eje y y color
axes[0][1].set_ylabel("Longitud del ptalo")
axes[0][1].set_xlabel("Longitud del spalo")

# Grfica 3
axes[0][2].scatter(sl, pw, c=iris['color']) # Asignacin de eje x, eje y y color
axes[0][2].set_ylabel("Ancho del ptalo")
axes[0][2].set_xlabel("Longitud del spalo")

# Grfica 4
axes[1][0].scatter(sw, pl, c=iris['color']) # Asignacin de eje x, eje y y color
axes[1][0].set_ylabel("Longitud del ptalo")
axes[1][0].set_xlabel("Ancho del spalo")

# Grfica 5
axes[1][1].scatter(sw, pw, c=iris['color']) # Asignacin de eje x, eje y y color
axes[1][1].set_ylabel("Ancho del ptalo")
axes[1][1].set_xlabel("Ancho del spalo")

# Grfica 6
axes[1][2].scatter(pw, pl, c=iris['color']) # Asignacin de eje x, eje y y color
axes[1][2].set_ylabel("Longitud del ptalo")
axes[1][2].set_xlabel("Ancho del ptalo")

plt.show()
```
