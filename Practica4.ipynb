```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,1,figsize=(5,5))
iris = pd.read_csv("./iris.csv")
print((iris.shape))
pw = iris["petal.width"]
pl = iris["petal.length"]
sw = iris["sepal.width"]
sl = iris["sepal.length"]

print(len(pw))

A = np.zeros((150,4)) # Matriz de 150 filas por 4 columnas
A[:,0] = pw # Se asignan todos los valores de pw a la columna 0
A[:,1] = pl
A[:,2] = sw
A[:,3] = sl

axes.scatter(pw, pl)
axes.set_ylabel("Longitud del pétalo")
axes.set_xlabel("Ancho del pétalo")
plt.show()
```
