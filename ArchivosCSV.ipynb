```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def distancia_euclidiana(x:list, y:list):
    return np.sqrt(np.sum((x - y) ** 2))

iris = pd.read_csv("./iris.csv")
matriz_distancias = np.zeros((150,150)) 
for i in range(150):
    for j in range(150):
        matriz_distancias[i,j] = distancia_euclidiana(iris.iloc[i,0:4], iris.iloc[j,0:4])
print(matriz_distancias)

fig, axes = plt.subplots(1,1,figsize=(10,10))
axes.imshow(matriz_distancias)

```
