```python
# Una red es una colección de nodos y enlaces
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 5))
#G = nx.Graph() # Grafo no dirigido
G = nx.DiGraph() # Grafo dirigido
"""
G.add_node(0,pos=[1,1]) # añade un nodo e indica la posición
G.add_node(1,pos=[1,3]) 
G.add_node(2,pos=[3,2]) 
G.add_node(3,pos=[4,4]) 
G.add_node(4,pos=[5,5]) 
"""
G.add_edge(2,4) # añade un enlace entre dos nodos
G.add_edge(2,3)
G.add_edge(0,3)
G.add_edge(1,2)
G.add_edge(1,3)
#pos = nx.get_node_attributes(G,'pos') # obtiene las posiciones de los nodos
pos = nx.circular_layout(G) # se indica la estructura para la visualización
nx.draw_networkx_nodes(G,pos=pos,nodelist=range(0,2),node_color='yellowgreen',node_size=[100])
nx.draw_networkx_nodes(G,pos=pos,nodelist=range(2,5),node_color='c',label='nodo')
nx.draw_networkx_edges(G,pos=pos,arrowsize=18,width=1.0,edge_color='navy') # muestra enlaces

nx.draw_networkx_labels(G,pos=pos) # muestra etiquetas
plt.legend()
plt.savefig('red.png',dpi=300)
```
