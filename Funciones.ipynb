Funciones

Una función es un bloque de código que realiza una tarea específica y puede usarse múltiples veces. Las funciones se definen con un nombre y pueden tener entradas y salidas.
En Python, se utiliza la palabra def seguida del nombre de la función, entre paréntesis las entradas y dos puntos ':' para indicar el inicio del bloque de código de la función.

def nombre_funcion(entradas):
    #cuerpo de la funcione
    return valor #opcional

Las funciones sirven para reducir el número de líneas de código.
Ejemplo: Hacer una función que devuelva el cuadrado de un número.
def cuadrado(x):
    return x*x
    #x**2
y = cuadrado(10)

Crear una función que regrese el factorial de un número:

```python
def factorial(num: int)->int:
    resultado = 1
    for i in range(1,num+1):
        resultado *= i
    return resultado

num = 5
print("El factorial iterativo de ", num, " es: ", factorial(num))

def factorial2(x: int)->int:
    if(x==0 or x==1):
        return 1
    else:
        return x*factorial2(x-1)

num = 5
print("El factorial recursivo de ", num, " es: ", factorial2(num))
```

```python
import numpy as np

def minimo_maximo_lista():
    nums = np.random.rand(10)
    nums.sort()
    print(nums)
    print("Maximo: ", nums[-1])
    print("Minimo: ", nums[0])

minimo_maximo_lista()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_matriz(d: int, color):
    fig, axes = plt.subplots(1, 1,figsize=(4,4))
    matrix = np.random.rand(d,d)
    axes.imshow(matrix,cmap = color)

graficar_matriz(5, 'Blues')
```

linspace es una función de la biblioteca NumPy que crea un arreglo de npumeros equiespaciados de un rango específico.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
```
