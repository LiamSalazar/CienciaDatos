```python
def matriz_identidad_bucles(n):
    matriz = []
    for i in range(n):
        fila = []
        for j in range(n):
            if i == j:
                fila.append(1)
            else:
                fila.append(0)
        matriz.append(fila)
    return matriz

def matriz_identidad_comprension(n):
    return [[1 if i == j else 0 for j in range(n)] for i in range(n)]

n = 3

print(f"Matriz con for de ({n}x{n}):")
for fila in matriz_identidad_bucles(n):
    print(fila)

print(f"Matriz con compresión de ({n}x{n}):")
for fila in matriz_identidad_comprension(n):
    print(fila)
```
