```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations

# Ajustes globales
plt.rcParams["figure.figsize"] = (7.5, 5.0)
plt.rcParams["axes.grid"] = True

DATA_PATH = "population_data.csv"

# Límites base de cuantiles y relajación
Q_LOW_BASE  = 0.01
Q_HIGH_BASE = 0.97
MIN_KEEP_RATIO = 0.95  
RELAX_STEP = 0.02       

# Escala para gráficas ("none","x","y","xy","auto")
SCATTER_SCALE_MODE = "auto"
PDF_SCALE_MODE     = "none"

# Umbrales simples para sugerir log en PDFs
PDF_LOG_SPAN_THRESHOLD = 100.0  
PDF_MIN_POSITIVE = 1e-12        

# Utilidades
def pick_country_column(df: pd.DataFrame) -> str | None:
    for col in df.columns:
        if "country" in col.lower():
            return col
    return None

def safe_log(arr: np.ndarray) -> np.ndarray:
    """log natural con recorte a positivo mínimo."""
    a = arr.astype(float)
    return np.log(np.clip(a, PDF_MIN_POSITIVE, None))

def quantile_info(arr: np.ndarray, q_low: float, q_high: float):
    a = arr.astype(float)
    finite = np.isfinite(a)
    if finite.sum() == 0:
        return finite, np.nan, np.nan
    ql = np.nanquantile(a[finite], q_low)
    qh = np.nanquantile(a[finite], q_high)
    mask = finite & (a >= ql) & (a <= qh)
    return mask, float(ql), float(qh)

def adaptive_mask(arr: np.ndarray,
                  q_low: float = Q_LOW_BASE,
                  q_high: float = Q_HIGH_BASE,
                  min_keep_ratio: float = MIN_KEEP_RATIO,
                  step: float = RELAX_STEP):
    """Relaja cuantiles hasta retener al menos min_keep_ratio."""
    a = arr.astype(float)
    finite = np.isfinite(a)
    if finite.sum() == 0:
        return finite, np.nan, np.nan, q_low, q_high
    n = finite.sum()
    cur_low, cur_high = q_low, q_high
    while True:
        m, ql, qh = quantile_info(a, cur_low, cur_high)
        keep_ratio = (m & finite).sum() / n
        if keep_ratio >= min_keep_ratio or (cur_low <= 0.0 and cur_high >= 1.0):
            return m, ql, qh, cur_low, cur_high
        cur_low = max(0.0, cur_low - step)
        cur_high = min(1.0, cur_high + step)

def joint_mask(x: np.ndarray, y: np.ndarray,
               q_low: float = Q_LOW_BASE, q_high: float = Q_HIGH_BASE,
               min_keep_ratio: float = MIN_KEEP_RATIO, step: float = RELAX_STEP):
    """Máscara conjunta: filtra por cuantiles (relajados) en cada eje y combina."""
    mx, qlx, qhx, eff_lx, eff_hx = adaptive_mask(x, q_low, q_high, min_keep_ratio, step)
    my, qly, qhy, eff_ly, eff_hy = adaptive_mask(y, q_low, q_high, min_keep_ratio, step)
    m = mx & my
    info = {
        "x_limits": (qlx, qhx),
        "y_limits": (qly, qhy),
        "x_q_used": (eff_lx, eff_hx),
        "y_q_used": (eff_ly, eff_hy),
        "x_kept": int(mx.sum()),
        "y_kept": int(my.sum())
    }
    return m, info

def choose_best_scale_for_scatter(x: np.ndarray, y: np.ndarray, mask: np.ndarray) -> str:
    """
    Elige entre none/x/y/xy comparando R² de regresión lineal
    en (x,y), (log x, y), (x, log y), (log x, log y) usando datos filtrados y positivos donde aplique.
    """

    def r2_of(u, v):
        a0, a1, r2 = StatsUtils.linear_regression(u, v)
        return r2 if np.isfinite(r2) else -np.inf

    xf, yf = x[mask], y[mask]
    modes = ["none", "x", "y", "xy"]
    scores = {}

    # NONE
    scores["none"] = r2_of(xf, yf)

    # X log
    if np.all(xf > 0):
        scores["x"] = r2_of(safe_log(xf), yf)
    else:
        scores["x"] = -np.inf

    # Y log
    if np.all(yf > 0):
        scores["y"] = r2_of(xf, safe_log(yf))
    else:
        scores["y"] = -np.inf

    # XY log-log
    if np.all(xf > 0) and np.all(yf > 0):
        scores["xy"] = r2_of(safe_log(xf), safe_log(yf))
    else:
        scores["xy"] = -np.inf

    # el mejor
    best = max(modes, key=lambda m: scores[m])
    return best

def apply_axis_scale(ax: plt.Axes, mode: str):
    if mode in ("x", "xy"):
        ax.set_xscale("log")
    if mode in ("y", "xy"):
        ax.set_yscale("log")

# Utilidades estadísticas
class StatsUtils:

    @staticmethod
    def mean(x: np.ndarray) -> float:
        return float(np.sum(x) / len(x)) if len(x) > 0 else np.nan

    @staticmethod
    def var(x: np.ndarray) -> float:
        if len(x) <= 1:
            return np.nan
        mu = StatsUtils.mean(x)
        return float(np.sum((x - mu) ** 2) / (len(x) - 1))

    @staticmethod
    def std(x: np.ndarray) -> float:
        v = StatsUtils.var(x)
        return float(np.sqrt(v)) if np.isfinite(v) else np.nan

    @staticmethod
    def cov(x: np.ndarray, y: np.ndarray) -> float:
        n = len(x)
        if n <= 1 or len(y) != n:
            return np.nan
        mx, my = StatsUtils.mean(x), StatsUtils.mean(y)
        return float(np.sum((x - mx) * (y - my)) / (n - 1))

    @staticmethod
    def corr(x: np.ndarray, y: np.ndarray) -> float:
        vx, vy = StatsUtils.var(x), StatsUtils.var(y)
        if not np.isfinite(vx) or not np.isfinite(vy) or vx == 0 or vy == 0:
            return np.nan
        return StatsUtils.cov(x, y) / (np.sqrt(vx) * np.sqrt(vy))

    @staticmethod
    def linear_regression(x: np.ndarray, y: np.ndarray) -> tuple[float, float, float]:
        n = len(x)
        if n == 0 or n != len(y):
            return np.nan, np.nan, np.nan
        mx, my = StatsUtils.mean(x), StatsUtils.mean(y)
        sxx = float(np.sum((x - mx) ** 2))
        sxy = float(np.sum((x - mx) * (y - my)))
        if sxx == 0:
            ss_tot = float(np.sum((y - my) ** 2))
            r2 = 0.0 if ss_tot > 0 else np.nan
            return float(my), np.nan, r2
        a1 = sxy / sxx
        a0 = my - a1 * mx
        y_hat = a0 + a1 * x
        ss_res = float(np.sum((y - y_hat) ** 2))
        ss_tot = float(np.sum((y - my) ** 2))
        r2 = 0.0 if ss_tot == 0 else 1.0 - ss_res / ss_tot
        return float(a0), float(a1), float(r2)

# Estimador de PDF con reporte y escala
class PDFEstimator:

    def __init__(self, bins: int | None = None):
        self.bins = bins

    @staticmethod
    def _auto_bins(n: int) -> int:
        return int(np.ceil(np.sqrt(n))) if n > 0 else 1

    def estimate(self, x: np.ndarray):
        # filtra con cuantiles adaptativos
        mask, ql, qh, eff_l, eff_h = adaptive_mask(x)
        data = x[mask]
        # decidir escala (auto o fija)
        scale_mode = PDF_SCALE_MODE
        if scale_mode == "auto":
            if np.all(data > 0) and (np.nanmax(data) / max(np.nanmin(data), PDF_MIN_POSITIVE) > PDF_LOG_SPAN_THRESHOLD):
                scale_mode = "x"
            else:
                scale_mode = "none"

        # histograma según escala
        if scale_mode == "x" and np.all(data > 0):
            # bins log-uniformes
            lo = max(np.nanmin(data), PDF_MIN_POSITIVE)
            hi = np.nanmax(data)
            n_bins = self.bins or self._auto_bins(len(data))
            edges = np.geomspace(lo, hi, num=n_bins + 1)
            counts, edges = np.histogram(data, bins=edges, density=True)
        else:
            n_bins = self.bins or self._auto_bins(len(data))
            counts, edges = np.histogram(data, bins=n_bins, density=True)

        centers = 0.5 * (edges[:-1] + edges[1:])
        info = {"ql": ql, "qh": qh, "q_used": (eff_l, eff_h), "kept": int(mask.sum()), "scale": scale_mode}
        return centers, counts, mask, info

    def plot(self, series: pd.Series, df: pd.DataFrame | None = None, country_col: str | None = None, title: str | None = None) -> None:
        values = series.values.astype(float)
        centers, pdf, mask, info = self.estimate(values)

        fig, ax = plt.subplots()
        width = (centers[1] - centers[0]) if len(centers) > 1 else 1.0
        ax.bar(centers, pdf, width=width)
        ax.set_xlabel(series.name)
        ax.set_ylabel("Densidad aprox.")
        ax.set_yscale("log")
        ax.set_title(title or f"Distribución aproximada de {series.name}")

        # aplicar escala en eje X si info["scale"] == "x"
        apply_axis_scale(ax, "x" if info.get("scale", "none") == "x" else "none")
        plt.show()

        # Reporte de excluidos
        if df is not None and country_col is not None:
            excl_idx = series.index[~mask]
            excl_names = df.loc[excl_idx, country_col].astype(str).tolist()
            if excl_names:
                ql, qh = info["ql"], info["qh"]
                used_l, used_h = info["q_used"]
                print(
                    f"[PDF] Excluidos en '{series.name}' por cuantiles "
                    f"[{used_l:.2f}, {used_h:.2f}] (límites: {ql:.4g}–{qh:.4g}) — n={len(excl_names)}"
                )
                print("  - " + ", ".join(excl_names))

# Analizador del dataset
class DatasetAnalyzer:

    def __init__(self, df: pd.DataFrame):
        self.df = df.copy()
        self.numeric_columns = [
            c for c in self.df.columns if np.issubdtype(self.df[c].dtype, np.number)
        ]

    def correlation_matrix(self) -> pd.DataFrame:
        cols = self.numeric_columns
        corr_mat = np.full((len(cols), len(cols)), np.nan, dtype=float)
        for i, ci in enumerate(cols):
            xi = self.df[ci].values.astype(float)
            for j, cj in enumerate(cols):
                xj = self.df[cj].values.astype(float)
                mx, _, _, _, _ = adaptive_mask(xi)
                my, _, _, _, _ = adaptive_mask(xj)
                m = mx & my
                if m.sum() >= 2:
                    corr_mat[i, j] = StatsUtils.corr(xi[m], xj[m])
                else:
                    corr_mat[i, j] = np.nan
        return pd.DataFrame(corr_mat, index=cols, columns=cols)

    def linear_regression_pairs(self) -> pd.DataFrame:
        rows = []
        for x_name, y_name in combinations(self.numeric_columns, 2):
            x = self.df[x_name].values.astype(float)
            y = self.df[y_name].values.astype(float)
            m, _ = joint_mask(x, y)
            if m.sum() >= 2:
                a0, a1, r2 = StatsUtils.linear_regression(x[m], y[m])
            else:
                a0, a1, r2 = np.nan, np.nan, np.nan
            rows.append({"x": x_name, "y": y_name, "a0": a0, "a1": a1, "R2": r2})

            if m.sum() >= 2:
                a0b, a1b, r2b = StatsUtils.linear_regression(y[m], x[m])
            else:
                a0b, a1b, r2b = np.nan, np.nan, np.nan
            rows.append({"x": y_name, "y": x_name, "a0": a0b, "a1": a1b, "R2": r2b})

        out = pd.DataFrame(rows)
        if "R2" in out.columns:
            out["R2"] = out["R2"].clip(lower=0.0, upper=1.0)
        return out

    def sum_R2(self) -> float:
        reg = self.linear_regression_pairs()
        if "R2" not in reg.columns or reg["R2"].dropna().empty:
            return float("nan")
        return float(reg["R2"].dropna().sum())

    def scatter_with_regression(self, x_name: str, y_name: str, show_regression: bool = True) -> None:
        x = self.df[x_name].values.astype(float)
        y = self.df[y_name].values.astype(float)

        base_mask, info = joint_mask(x, y)
        xf, yf = x[base_mask], y[base_mask]

        # Etiquetas por continente
        if "Continent" in self.df.columns:
            labels_all = self.df["Continent"].astype(str).values
        else:
            labels_all = np.array(["Unknown"] * len(self.df))
        labels = labels_all[base_mask]

        # Elegir escala
        if SCATTER_SCALE_MODE == "auto":
            scale_mode = choose_best_scale_for_scatter(x, y, base_mask)
        else:
            scale_mode = SCATTER_SCALE_MODE

        # Preparar datos transformados según escala para la línea de regresión
        if scale_mode == "none":
            xs_r, ys_r = xf, yf
        elif scale_mode == "x":
            xs_r, ys_r = safe_log(xf), yf
        elif scale_mode == "y":
            xs_r, ys_r = xf, safe_log(yf)
        elif scale_mode == "xy":
            xs_r, ys_r = safe_log(xf), safe_log(yf)
        else:
            xs_r, ys_r = xf, yf  # fallback

        # Reporte de excluidos (motivos)
        country_col = pick_country_column(self.df)
        if country_col is not None:
            mx, qlx, qhx, _, _ = adaptive_mask(x)
            my, qly, qhy, _, _ = adaptive_mask(y)
            m_joint = mx & my
            excl_rows = []
            for i in np.where(~m_joint)[0]:
                reasons = []
                if not mx[i]:
                    if np.isfinite(x[i]):
                        if x[i] < qlx: reasons.append(f"{x_name}<{qlx:.4g}")
                        elif x[i] > qhx: reasons.append(f"{x_name}>{qhx:.4g}")
                    else:
                        reasons.append(f"{x_name}=NaN/Inf")
                if not my[i]:
                    if np.isfinite(y[i]):
                        if y[i] < qly: reasons.append(f"{y_name}<{qly:.4g}")
                        elif y[i] > qhy: reasons.append(f"{y_name}>{qhy:.4g}")
                    else:
                        reasons.append(f"{y_name}=NaN/Inf")
                if reasons:
                    excl_rows.append(f"{self.df.iloc[i][country_col]} ({'; '.join(reasons)})")
            if excl_rows:
                used_x_q = info["x_q_used"]; used_y_q = info["y_q_used"]
                lim_x = info["x_limits"]; lim_y = info["y_limits"]
                print(
                    f"[SCATTER] Excluidos en '{x_name} vs {y_name}' "
                    f"[x q={used_x_q}, lim={lim_x}; y q={used_y_q}, lim={lim_y}] — n={len(excl_rows)}"
                )
                for r in excl_rows:
                    print("  - " + r)

        # Gráfica
        fig, ax = plt.subplots()
        for lab in np.unique(labels):
            m = (labels == lab)
            ax.scatter(xf[m], yf[m], label=str(lab))
        ax.set_xlabel(x_name)
        ax.set_ylabel(y_name)
        ax.set_title(f"Dispersión: {x_name} vs {y_name}")
        apply_axis_scale(ax, scale_mode)  # aplica escala visual
        ax.legend()

        # Regresión en el espacio transformado elegido
        if show_regression and len(xs_r) > 1:
            a0, a1, r2 = StatsUtils.linear_regression(xs_r, ys_r)
            xr_line = np.linspace(np.nanmin(xs_r), np.nanmax(xs_r), 100)
            yr_line = a0 + a1 * xr_line

            # dibujar la línea en el mismo espacio transformado
            if scale_mode == "none":
                ax.plot(xr_line, yr_line, linewidth=2)
                ax.set_title(f"{x_name} vs {y_name} — y = {a0:.3f} + {a1:.3f}x | R² = {r2:.3f}")
            elif scale_mode == "x":
                # xr_line = log(x) -> línea en log-x; el eje ya está en log, pero datos originales están en x
                # Para mostrar coherente, convertimos xr_line a x_line = exp(xr_line)
                x_line = np.exp(xr_line)
                ax.plot(x_line, yr_line, linewidth=2)
                ax.set_title(f"{x_name} (log) vs {y_name} — y = {a0:.3f} + {a1:.3f}log(x) | R² = {r2:.3f}")
            elif scale_mode == "y":
                # yr_line en espacio log(y): mostramos como y_line = exp(yr_line)
                y_line = np.exp(yr_line)
                ax.plot(xr_line, y_line, linewidth=2)
                ax.set_title(f"{x_name} vs {y_name} (log) — log(y) = {a0:.3f} + {a1:.3f}x | R² = {r2:.3f}")
            elif scale_mode == "xy":
                # xr_line, yr_line en log-log: x=exp(xr), y=exp(yr)
                x_line = np.exp(xr_line)
                y_line = np.exp(yr_line)
                ax.plot(x_line, y_line, linewidth=2)
                ax.set_title(f"{x_name} (log) vs {y_name} (log) — log(y) = {a0:.3f} + {a1:.3f}log(x) | R² = {r2:.3f}")

        plt.show()

    def plot_all_pdfs(self) -> None:
        estimator = PDFEstimator()
        country_col = pick_country_column(self.df)
        for col in self.numeric_columns:
            estimator.plot(self.df[col], df=self.df, country_col=country_col,
                           title=f"Distribución aproximada de {col}")

    def plot_all_scatters(self, with_regression: bool = True) -> None:
        for x_name, y_name in combinations(self.numeric_columns, 2):
            self.scatter_with_regression(x_name, y_name, show_regression=with_regression)

# Carga y ejecución
df = pd.read_csv(DATA_PATH)

# Quitar columnas irrelevantes
if "id" in df.columns:
    df = df.drop(columns=["id"])

for col in ["World Share", "WorldShare", "World_Share"]:
    if col in df.columns:
        df = df.drop(columns=[col])

analyzer = DatasetAnalyzer(df)

# Matriz de correlación
corr_df = analyzer.correlation_matrix()
print("\n MATRIZ DE CORRELACIÓN DE PEARSON ")
display(corr_df.round(4))

# PDFs
analyzer.plot_all_pdfs()

# Dispersión + regresión (con escala auto)
analyzer.plot_all_scatters(with_regression=True)

# Coeficientes y R^2
reg_table = analyzer.linear_regression_pairs().round(6)
print("COEFICIENTES Y R^2 (TODOS LOS PARES, AMBAS DIRECCIONES)")
display(reg_table)

total_r2 = analyzer.sum_R2()
print(f"\nSumatoria total de R^2 en todas las direcciones: {total_r2:.6f}")
```
