```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations

# Ajustes globales para gráficos
plt.rcParams["figure.figsize"] = (7.5, 5.0)
plt.rcParams["axes.grid"] = True

DATA_PATH = "population_data.csv"  

# Funciones de estadpística
class StatsUtils:

    @staticmethod
    def mean(x: np.ndarray) -> float: # Promedio
        return float(np.sum(x) / len(x))

    @staticmethod
    def var(x: np.ndarray) -> float: # Varianza
        mu = StatsUtils.mean(x)
        return float(np.sum((x - mu) ** 2) / (len(x) - 1))

    @staticmethod
    def std(x: np.ndarray) -> float: # Desviación estándar
        return float(np.sqrt(StatsUtils.var(x)))

    @staticmethod
    def cov(x: np.ndarray, y: np.ndarray) -> float: # Covarianza
        mx, my = StatsUtils.mean(x), StatsUtils.mean(y)
        return float(np.sum((x - mx) * (y - my)) / (len(x) - 1))

    @staticmethod
    def corr(x: np.ndarray, y: np.ndarray) -> float: # Correlación de Pearson
        vx, vy = StatsUtils.var(x), StatsUtils.var(y)
        if vx == 0 or vy == 0:
            return np.nan
        return StatsUtils.cov(x, y) / (np.sqrt(vx) * np.sqrt(vy))

    @staticmethod
    def linear_regression(x: np.ndarray, y: np.ndarray) -> tuple[float, float, float]: # Regresión lineal (MCO)
        mx, my = StatsUtils.mean(x), StatsUtils.mean(y) 
        # Suma de cuadrados
        sxx = float(np.sum((x - mx) ** 2)) 
        sxy = float(np.sum((x - mx) * (y - my)))

        if sxx == 0: # Pendiente cero que da R^2 = 0
            return float(my), np.nan, 0.0

        a1 = sxy / sxx
        a0 = my - a1 * mx

        y_hat = a0 + a1 * x
        ss_res = float(np.sum((y - y_hat) ** 2))
        ss_tot = float(np.sum((y - my) ** 2))
        r2 = 0.0 if ss_tot == 0 else 1.0 - ss_res / ss_tot
        return float(a0), float(a1), float(r2)

# Estimador de distribución de probabilidad con histograma
class PDFEstimator:

    def __init__(self, bins: int | None = None):
        self.bins = bins

    @staticmethod 
    def _auto_bins(n: int) -> int: # Generación de número de bins
        return int(np.ceil(np.sqrt(n))) if n > 0 else 1

    def estimate(self, x: np.ndarray) -> tuple[np.ndarray, np.ndarray]: # Estimación de función de densidad de probabilidad
        data = x[~np.isnan(x)]
        n_bins = self.bins or self._auto_bins(len(data))
        counts, edges = np.histogram(data, bins=n_bins, density=True)
        centers = 0.5 * (edges[:-1] + edges[1:])
        return centers, counts

    def plot(self, series: pd.Series, title: str| None = None) -> None:
        values = series.values.astype(float)
        values = values[np.isfinite(values) & (values > 0)]
        centers, pdf = self.estimate(values)
        width = (centers[1] - centers[0]) if len(centers) > 1 else 1.0
        plt.figure()
        plt.bar(centers, pdf, width=width)
        plt.xlabel(series.name)
        plt.ylabel("Densidad aprox.")
        plt.title(title or f"Distribución aproximada de {series.name}")
        plt.show()

# Analizar el dataset
class DatasetAnalyzer:

    def __init__(self, df: pd.DataFrame, class_column: str| None = None): # Constructor
        self.df = df.copy()
        self.class_column = class_column
        self.numeric_columns = [
            c for c in self.df.columns if np.issubdtype(self.df[c].dtype, np.number) # Filtrar columnas numéricas
        ]

    def correlation_matrix(self) -> pd.DataFrame:
        cols = self.numeric_columns
        corr_mat = np.full((len(cols), len(cols)), np.nan, dtype=float)
        for i, ci in enumerate(cols):
            xi = self.df[ci].values.astype(float)
            for j, cj in enumerate(cols):
                xj = self.df[cj].values.astype(float)
                corr_mat[i, j] = StatsUtils.corr(xi, xj)
        return pd.DataFrame(corr_mat, index=cols, columns=cols)

    def linear_regression_pairs(self) -> pd.DataFrame:
        rows = []
        for x_name, y_name in combinations(self.numeric_columns, 2):
            x = self.df[x_name].values.astype(float)
            y = self.df[y_name].values.astype(float)
            a0, a1, r2 = StatsUtils.linear_regression(x, y)
            rows.append({"x": x_name, "y": y_name, "a0": a0, "a1": a1, "R2": r2})
            # Regresión inversa
            a0b, a1b, r2b = StatsUtils.linear_regression(y, x)
            rows.append({"x": y_name, "y": x_name, "a0": a0b, "a1": a1b, "R2": r2b})

        out = pd.DataFrame(rows)
        out["R2"] = out["R2"].clip(lower=0.0, upper=1.0)
        return out

    def sum_R2(self) -> float: # Sumatoria de errores para cada una de las parejas
        reg = self.linear_regression_pairs()
        return float(reg["R2"].sum())

    def _pick_or_build_classes(self, x_name: str, y_name: str) -> pd.Series:
        if False and (
            self.class_column is not None
            and self.class_column in self.df.columns
            and self.df[self.class_column].nunique() <= 10
        ):
            return self.df[self.class_column].astype(str)

        y = self.df[y_name].values.astype(float)
        y_clean = pd.Series(y).dropna()
        q1, q2 = y_clean.quantile([0.33, 0.66])
        def label_y(val: float) -> str:
            if np.isnan(val):
                return "sin datos"
            if val <= q1:
                return "bajo"
            if val <= q2:
                return "medio"
            return "alto"

        return pd.Series([label_y(v) for v in y], index=self.df.index, name="class")

    def scatter_with_regression( # Gráfica de dispersión con recta de regresión
        self, x_name: str, y_name: str, show_regression: bool = True
    ) -> None:
        x = self.df[x_name].values.astype(float)
        y = self.df[y_name].values.astype(float)

        # Clases para colorear dependiendo si es alto, medio o bajo o ya está la categoría
        class_labels = self._pick_or_build_classes(x_name, y_name)

        # Graficar por grupo de clase
        plt.figure()
        for label, group_idx in class_labels.groupby(class_labels).groups.items():
            plt.scatter(x[group_idx], y[group_idx], label=str(label))
        plt.xlabel(x_name)
        plt.ylabel(y_name)
        plt.title(f"Dispersión: {x_name} vs {y_name}")
        plt.legend()

        # Regresión lineal
        if show_regression:
            a0, a1, r2 = StatsUtils.linear_regression(x, y)
            x_line = np.linspace(np.nanmin(x), np.nanmax(x), 100)
            y_line = a0 + a1 * x_line
            plt.plot(x_line, y_line, linewidth=2)
            plt.title(
                f"{x_name} vs {y_name} (Regresión) — y = {a0:.3f} + {a1:.3f}x | R² = {r2:.3f}"
            )

        plt.show()

    def plot_all_pdfs(self) -> None: # Gráficas de distribución de probabilidad
        estimator = PDFEstimator()
        for col in self.numeric_columns:
            estimator.plot(self.df[col], title=f"Distribución aproximada de {col}")

    def plot_all_scatters(self, with_regression: bool = True) -> None: # Gráficas de dispersión con regresión
        for x_name, y_name in combinations(self.numeric_columns, 2):
            self.scatter_with_regression(x_name, y_name, show_regression=with_regression)

df = pd.read_csv(DATA_PATH)

# Quitar los ID's porque no son relevantes en el análisis
if "id" in df.columns:
    df = df.drop(columns=["id"])

# Detectar una posible columna categórica (pocas categorías)
possible_class_col = None
for col in df.columns:
    is_numeric = np.issubdtype(df[col].dtype, np.number)
    nunique = df[col].nunique(dropna=True)
    if (not is_numeric and nunique <= 10) or (is_numeric and nunique <= 10 and nunique > 1):
        possible_class_col = col
        break

analyzer = DatasetAnalyzer(df, class_column=possible_class_col)

corr_df = analyzer.correlation_matrix()
print("\n MATRIZ DE CORRELACIÓN DE PEARSON ")
display(corr_df.round(4))

# Distribuciones de probabilidad
analyzer.plot_all_pdfs()

# Gráficas de dispersión con la regresión lineal
analyzer.plot_all_scatters(with_regression=True)

#Coeficientes y error en regresión lineal
reg_table = analyzer.linear_regression_pairs().round(6)
print("COEFICIENTES Y R^2 (TODOS LOS PARES, AMBAS DIRECCIONES)")
display(reg_table)

total_r2 = analyzer.sum_R2()
print(f"\nSumatoria total de R^2 en todas las direcciones: {total_r2:.6f}")
```
