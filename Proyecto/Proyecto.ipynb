```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations

# Ajustes globales para gráficos
plt.rcParams["figure.figsize"] = (7.5, 5.0)
plt.rcParams["axes.grid"] = True

DATA_PATH = "population_data.csv"

# LÍMITES DE CUANTILES 
Q_LOW_BASE  = 0.01
Q_HIGH_BASE = 0.97
MIN_KEEP_RATIO = 0.95   
RELAX_STEP = 0.01       

# Utilidades: país / columnas
def pick_country_column(df: pd.DataFrame) -> str | None:
    for col in df.columns:
        if "country" in col.lower():
            return col
    return None

# Utilidades de cuantiles
def quantile_info(arr: np.ndarray, q_low: float, q_high: float):
    a = arr.astype(float)
    finite = np.isfinite(a)
    if finite.sum() == 0:
        return finite, np.nan, np.nan
    ql = np.nanquantile(a[finite], q_low)
    qh = np.nanquantile(a[finite], q_high)
    mask = finite & (a >= ql) & (a <= qh)
    return mask, float(ql), float(qh)

def adaptive_mask(arr: np.ndarray,
                  q_low: float = Q_LOW_BASE,
                  q_high: float = Q_HIGH_BASE,
                  min_keep_ratio: float = MIN_KEEP_RATIO,
                  step: float = RELAX_STEP):
    """Relaja los cuantiles hasta retener al menos min_keep_ratio de datos."""
    a = arr.astype(float)
    finite = np.isfinite(a)
    if finite.sum() == 0:
        return finite, np.nan, np.nan, q_low, q_high
    n = finite.sum()
    cur_low, cur_high = q_low, q_high
    while True:
        m, ql, qh = quantile_info(a, cur_low, cur_high)
        keep_ratio = (m & finite).sum() / n
        if keep_ratio >= min_keep_ratio or (cur_low <= 0.0 and cur_high >= 1.0):
            return m, ql, qh, cur_low, cur_high
        # relajar: bajar q_low y subir q_high
        cur_low = max(0.0, cur_low - step)
        cur_high = min(1.0, cur_high + step)

def joint_mask(x: np.ndarray, y: np.ndarray,
               q_low: float = Q_LOW_BASE, q_high: float = Q_HIGH_BASE,
               min_keep_ratio: float = MIN_KEEP_RATIO, step: float = RELAX_STEP):
    """Máscara conjunta relajando por variable y luego combinando."""
    mx, qlx, qhx, eff_lx, eff_hx = adaptive_mask(x, q_low, q_high, min_keep_ratio, step)
    my, qly, qhy, eff_ly, eff_hy = adaptive_mask(y, q_low, q_high, min_keep_ratio, step)
    m = mx & my
    info = {
        "x_limits": (qlx, qhx),
        "y_limits": (qly, qhy),
        "x_q_used": (eff_lx, eff_hx),
        "y_q_used": (eff_ly, eff_hy),
        "x_kept": int(mx.sum()),
        "y_kept": int(my.sum())
    }
    return m, info

# -------------------------------
# Utilidades estadísticas
# -------------------------------
class StatsUtils:

    @staticmethod
    def mean(x: np.ndarray) -> float:  # Promedio
        return float(np.sum(x) / len(x)) if len(x) > 0 else np.nan

    @staticmethod
    def var(x: np.ndarray) -> float:  # Varianza
        if len(x) <= 1:
            return np.nan
        mu = StatsUtils.mean(x)
        return float(np.sum((x - mu) ** 2) / (len(x) - 1))

    @staticmethod
    def std(x: np.ndarray) -> float:  # Desviación estándar
        v = StatsUtils.var(x)
        return float(np.sqrt(v)) if np.isfinite(v) else np.nan

    @staticmethod
    def cov(x: np.ndarray, y: np.ndarray) -> float:  # Covarianza
        n = len(x)
        if n <= 1 or len(y) != n:
            return np.nan
        mx, my = StatsUtils.mean(x), StatsUtils.mean(y)
        return float(np.sum((x - mx) * (y - my)) / (n - 1))

    @staticmethod
    def corr(x: np.ndarray, y: np.ndarray) -> float:  # Correlación de Pearson
        vx, vy = StatsUtils.var(x), StatsUtils.var(y)
        if not np.isfinite(vx) or not np.isfinite(vy) or vx == 0 or vy == 0:
            return np.nan
        return StatsUtils.cov(x, y) / (np.sqrt(vx) * np.sqrt(vy))

    @staticmethod
    def linear_regression(x: np.ndarray, y: np.ndarray) -> tuple[float, float, float]:  # Regresión lineal (MCO)
        n = len(x)
        if n == 0 or n != len(y):
            return np.nan, np.nan, np.nan
        mx, my = StatsUtils.mean(x), StatsUtils.mean(y)
        sxx = float(np.sum((x - mx) ** 2))
        sxy = float(np.sum((x - mx) * (y - my)))
        if sxx == 0:  # Pendiente no definida
            # y = my (línea horizontal), R^2 = 0 si hay variación en y; NaN si no hay datos
            ss_tot = float(np.sum((y - my) ** 2))
            r2 = 0.0 if ss_tot > 0 else np.nan
            return float(my), np.nan, r2
        a1 = sxy / sxx
        a0 = my - a1 * mx
        y_hat = a0 + a1 * x
        ss_res = float(np.sum((y - y_hat) ** 2))
        ss_tot = float(np.sum((y - my) ** 2))
        r2 = 0.0 if ss_tot == 0 else 1.0 - ss_res / ss_tot
        return float(a0), float(a1), float(r2)

# -------------------------------
# Estimador de PDF con reporte
# -------------------------------
class PDFEstimator:

    def __init__(self, bins: int | None = None):
        self.bins = bins

    @staticmethod
    def _auto_bins(n: int) -> int:  # Generación de número de bins
        return int(np.ceil(np.sqrt(n))) if n > 0 else 1

    # devuelve también mask y límites
    def estimate(self, x: np.ndarray):
        mask, ql, qh, eff_l, eff_h = adaptive_mask(x)
        data = x[mask]
        n_bins = self.bins or self._auto_bins(len(data))
        counts, edges = np.histogram(data, bins=n_bins, density=True)
        centers = 0.5 * (edges[:-1] + edges[1:])
        info = {"ql": ql, "qh": qh, "q_used": (eff_l, eff_h), "kept": int(mask.sum())}
        return centers, counts, mask, info

    def plot(self, series: pd.Series, df: pd.DataFrame | None = None, country_col: str | None = None, title: str | None = None) -> None:
        values = series.values.astype(float)
        centers, pdf, mask, info = self.estimate(values)
        width = (centers[1] - centers[0]) if len(centers) > 1 else 1.0
        plt.figure()
        plt.bar(centers, pdf, width=width)
        plt.xlabel(series.name)
        plt.ylabel("Densidad aprox.")
        plt.title(title or f"Distribución aproximada de {series.name}")
        plt.show()

        # Reporte de excluidos
        if df is not None and country_col is not None:
            excl_idx = series.index[~mask]
            excl_names = df.loc[excl_idx, country_col].astype(str).tolist()
            if excl_names:
                ql, qh = info["ql"], info["qh"]
                used_l, used_h = info["q_used"]
                print(
                    f"[PDF] Excluidos en '{series.name}' por cuantiles "
                    f"[{used_l:.2f}, {used_h:.2f}] (límites: {ql:.4g}–{qh:.4g}) — n={len(excl_names)}"
                )
                print("  - " + ", ".join(excl_names))

# -------------------------------
# Analizador del dataset
# -------------------------------
class DatasetAnalyzer:

    def __init__(self, df: pd.DataFrame):  # Constructor
        self.df = df.copy()
        self.numeric_columns = [
            c for c in self.df.columns if np.issubdtype(self.df[c].dtype, np.number)
        ]

    def correlation_matrix(self) -> pd.DataFrame:
        cols = self.numeric_columns
        corr_mat = np.full((len(cols), len(cols)), np.nan, dtype=float)
        for i, ci in enumerate(cols):
            xi = self.df[ci].values.astype(float)
            for j, cj in enumerate(cols):
                xj = self.df[cj].values.astype(float)
                mx, _, _, _, _ = adaptive_mask(xi)
                my, _, _, _, _ = adaptive_mask(xj)
                m = mx & my
                if m.sum() >= 2:
                    corr_mat[i, j] = StatsUtils.corr(xi[m], xj[m])
                else:
                    corr_mat[i, j] = np.nan
        return pd.DataFrame(corr_mat, index=cols, columns=cols)

    def linear_regression_pairs(self) -> pd.DataFrame:
        rows = []
        for x_name, y_name in combinations(self.numeric_columns, 2):
            x = self.df[x_name].values.astype(float)
            y = self.df[y_name].values.astype(float)
            m, _ = joint_mask(x, y)
            if m.sum() >= 2:
                a0, a1, r2 = StatsUtils.linear_regression(x[m], y[m])
            else:
                a0, a1, r2 = np.nan, np.nan, np.nan
            rows.append({"x": x_name, "y": y_name, "a0": a0, "a1": a1, "R2": r2})

            # Regresión inversa
            if m.sum() >= 2:
                a0b, a1b, r2b = StatsUtils.linear_regression(y[m], x[m])
            else:
                a0b, a1b, r2b = np.nan, np.nan, np.nan
            rows.append({"x": y_name, "y": x_name, "a0": a0b, "a1": a1b, "R2": r2b})

        out = pd.DataFrame(rows)
        if "R2" in out.columns:
            out["R2"] = out["R2"].clip(lower=0.0, upper=1.0)
        return out

    def sum_R2(self) -> float:  # Sumatoria de R^2
        reg = self.linear_regression_pairs()
        if "R2" not in reg.columns or reg["R2"].dropna().empty:
            return float("nan")
        return float(reg["R2"].dropna().sum())

    def scatter_with_regression(self, x_name: str, y_name: str, show_regression: bool = True) -> None:
        x = self.df[x_name].values.astype(float)
        y = self.df[y_name].values.astype(float)

        # Máscara conjunta con relajación
        base_mask, info = joint_mask(x, y)

        xf, yf = x[base_mask], y[base_mask]

        # Etiquetas de clase = continente (filtradas)
        if "Continent" in self.df.columns:
            labels_all = self.df["Continent"].astype(str).values
        else:
            labels_all = np.array(["Unknown"] * len(self.df))
        labels = labels_all[base_mask]

        # Reporte de excluidos (con motivos)
        country_col = pick_country_column(self.df)
        if country_col is not None:
            mx, qlx, qhx, _, _ = adaptive_mask(x)
            my, qly, qhy, _, _ = adaptive_mask(y)
            m_joint = mx & my
            excl_rows = []
            for i in np.where(~m_joint)[0]:
                reasons = []
                if not mx[i]:
                    if np.isfinite(x[i]):
                        if x[i] < qlx: reasons.append(f"{x_name}<{qlx:.4g}")
                        elif x[i] > qhx: reasons.append(f"{x_name}>{qhx:.4g}")
                    else:
                        reasons.append(f"{x_name}=NaN/Inf")
                if not my[i]:
                    if np.isfinite(y[i]):
                        if y[i] < qly: reasons.append(f"{y_name}<{qly:.4g}")
                        elif y[i] > qhy: reasons.append(f"{y_name}>{qhy:.4g}")
                    else:
                        reasons.append(f"{y_name}=NaN/Inf")
                if reasons:
                    excl_rows.append(f"{self.df.iloc[i][country_col]} ({'; '.join(reasons)})")
            if excl_rows:
                used_x_q = info["x_q_used"]; used_y_q = info["y_q_used"]
                lim_x = info["x_limits"]; lim_y = info["y_limits"]
                print(
                    f"[SCATTER] Excluidos en '{x_name} vs {y_name}' "
                    f"[x q={used_x_q}, lim={lim_x}; y q={used_y_q}, lim={lim_y}] — n={len(excl_rows)}"
                )
                for r in excl_rows:
                    print("  - " + r)

        # Gráfica (con datos filtrados)
        plt.figure()
        for lab in np.unique(labels):
            m = (labels == lab)
            plt.scatter(xf[m], yf[m], label=str(lab))
        plt.xlabel(x_name)
        plt.ylabel(y_name)
        plt.title(f"Dispersión: {x_name} vs {y_name}")
        plt.legend()

        # Regresión con filtrados
        if show_regression and len(xf) > 1:
            a0, a1, r2 = StatsUtils.linear_regression(xf, yf)
            x_line = np.linspace(np.nanmin(xf), np.nanmax(xf), 100)
            y_line = a0 + a1 * x_line
            plt.plot(x_line, y_line, linewidth=2)
            plt.title(
                f"{x_name} vs {y_name} (Regresión) — y = {a0:.3f} + {a1:.3f}x | R² = {r2:.3f}"
            )

        plt.show()

    def plot_all_pdfs(self) -> None:
        estimator = PDFEstimator()
        country_col = pick_country_column(self.df)
        for col in self.numeric_columns:
            estimator.plot(self.df[col], df=self.df, country_col=country_col, title=f"Distribución aproximada de {col}")

    def plot_all_scatters(self, with_regression: bool = True) -> None:
        for x_name, y_name in combinations(self.numeric_columns, 2):
            self.scatter_with_regression(x_name, y_name, show_regression=with_regression)

# Ejecución
df = pd.read_csv(DATA_PATH)

# Quitar columnas irrelevantes
if "id" in df.columns:
    df = df.drop(columns=["id"])

for col in ["World Share", "WorldShare", "World_Share"]:
    if col in df.columns:
        df = df.drop(columns=[col])

analyzer = DatasetAnalyzer(df)

corr_df = analyzer.correlation_matrix()
print("\n MATRIZ DE CORRELACIÓN DE PEARSON ")
display(corr_df.round(4))

# Distribuciones de probabilidad
analyzer.plot_all_pdfs()

# Gráficas de dispersión con la regresión lineal
analyzer.plot_all_scatters(with_regression=True)

# Coeficientes y R^2 (todas las parejas, ambas direcciones)
reg_table = analyzer.linear_regression_pairs().round(6)
print("COEFICIENTES Y R^2 (TODOS LOS PARES, AMBAS DIRECCIONES)")
display(reg_table)

total_r2 = analyzer.sum_R2()
print(f"\nSumatoria total de R^2 en todas las direcciones: {total_r2:.6f}")
```
